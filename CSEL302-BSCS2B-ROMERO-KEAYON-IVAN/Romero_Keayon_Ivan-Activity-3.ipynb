```python
#@title Step 1: Importing Libraries
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt #Data Visualization
import numpy as np #Numerical Operation
from sklearn.linear_model import LinearRegression as lr #Data Modeling
```

```python
#@title Step 2: Loading Data Set
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head(1000)
```

```python
#@title Step 3: Visualizing the Data
plt.scatter(data['Previous_Exam_Scores'], data['Grades'], color='red')
plt.xlabel('Previous Exam Scores (X)')
plt.ylabel('Grades (Y)')
plt.title('Scatter Plot of Student Grades')
plt.show()
```

```python
#@title Step 4: Preparing Data for Modeling
X = data[['Previous_Exam_Scores']] #Independent Variable
Y = data[['Grades']] #Dependent Variable
```

```python
#@title Step 5: Fitting the Linear Regression Model
model = lr()

#Attributes of Linear Regression Function
model.fit(X, Y)
```

```python
#@title Step 6: Getting Model Coefficients
slope = model.coef_[0]

intercept = model.intercept_

print(f"Slope: {slope} \nIntercept: {intercept}")
```

```python
#@title Step 7: Visualizing the Linear Progression Line
plt.scatter(X, Y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Previous Exam Scores (X)')
plt.ylabel('Grades (Y)')
plt.title('Linear Progression of Student Grades')
```

```python
#@title Step 8: Making Predictions
new_X = [[np.random.randint(50, 100)]]

predicted_Y = model.predict(new_X)

print(f"Previous Exam Scores (X): {new_X} \nPredicted Grade (Y): {predicted_Y}")
```
