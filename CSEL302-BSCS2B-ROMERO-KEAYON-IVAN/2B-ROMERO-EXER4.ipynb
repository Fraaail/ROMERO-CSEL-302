```python
#@title Import Libraries
#@markdown - Import pandas, numpy, matplotlib, and sklearn.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
#@title 1. Data Exploration
#@markdown - In the data, there are five categories that can impact the grades of the students. Their grades will be based on hours of study, attendance, previous exam scores, extra curricular activities, and travel time to school.

data = pd.read_csv("Lesson_4_student_grades_dataset.csv")
data.head()
```

```python
#@title 2. Data Preprocessing
#@markdown - Standardizing the data of the students will help us to easily predict their grades based on the values of the independent variable given in the dataset.
X = data.drop('Grades', axis=1) #Independent Variables
Y = data[['Grades']] #Dependent Variable
```

```python
#@title 3. Train-Test Split
#@markdown - Splits the data.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#@title 4. Linear Regression Model
#@markdown - Linear Regression is used to predict the values of the student grade based on the coefficients of the independent variable.
model = LinearRegression()
```

```python
#@title 5. Model Training
#@markdown - Fit the data into the model.
model.fit(X_train, Y_train)
```

```python
#@title 7. Feature Importance Analysis
#@markdown - Each independent variable of the dataset will be combined in order to predict the grades of the student.
plt.scatter(data['Hours_of_Study'], Y)
plt.plot(data['Hours_of_Study'], model.predict(X), color='red')
plt.xlabel("Hours_of_Study")
plt.ylabel("Grades")
plt.title('Linear Regression of Data')
plt.show()

plt.scatter(data['Attendance_Percentage'], Y)
plt.plot(data['Attendance_Percentage'], model.predict(X), color='red')
plt.xlabel("Attendance_Percentage")
plt.ylabel("Grades")
plt.title('Scatter Plot of Data')
plt.show()

plt.scatter(data['Previous_Exam_Scores'], Y)
plt.plot(data['Previous_Exam_Scores'], model.predict(X), color='red')
plt.xlabel("Previous_Exam_Scores")
plt.ylabel("Grades")
plt.title('Scatter Plot of Data')
plt.show()

plt.scatter(data['Extracurricular_Activities'], Y)
plt.plot(data['Extracurricular_Activities'], model.predict(X), color='red')
plt.xlabel("Extracurricular_Activities")
plt.ylabel("Grades")
plt.title('Scatter Plot of Data')
plt.show()

plt.scatter(data['Travel_Time_to_School'], Y)
plt.plot(data['Travel_Time_to_School'], model.predict(X), color='red')
plt.xlabel("Travel_Time_to_School")
plt.ylabel("Grades")
plt.title('Scatter Plot of Data')
plt.show()
```

```python
#@title 8. Discussion
#@markdown - The independent variable that had the most impact on student grades is previous exam scores. This is due to the fact that it had the most accurate value of predicting the student grades.
#@markdown - Surprisingly, the attendance percentage of the students also had a high impact on predicting the grades of the students.
```
